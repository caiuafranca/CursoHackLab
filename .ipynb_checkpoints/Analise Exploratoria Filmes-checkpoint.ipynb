<h1>Filmes brasileiros e estrangeiros</h1>
url: https://oca.ancine.gov.br/cinema

```python
import pandas as pd
%matplotlib inline
```

```python
#skiprows e skipfooter limpa o cabeçalho e o footer dos arquivos
df = pd.read_excel('filmes_brasileiros_estrangeiros.xlsx',skiprows=2,skipfooter=16)
```

```python
df.head(20)
```

```python
df.dtypes
```

```python
#converter o campo Data de Lançamento em DATE
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
##Filmes com a data de Relançamento
df[df['Data de lançamento'] == 'Relançamento']

df_transformado = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
#Transformando Data Lançameto em DATE
df_transformado['Data de lançamento'] = pd.to_datetime(df_transformado['Data de lançamento'])
```

```python
df_transformado['Data de lançamento'].dt.year.value_counts()
```

```python
#Grafico de barras 
df_transformado['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
#Selecionando o Filme com a Maior Renda de Publico
df_transformado[df_transformado['Renda (R$) no ano de exibição'] == df['Renda (R$) no ano de exibição'].max()]
```
