```python
import pandas as pd
import requests
%matplotlib inline
```

```python
data = pd.read_csv('indeferidos-compet-12-2018.csv', sep=';', encoding='cp1252')
```

```python
dados_rn = data[data['UF'] == 'Rio Grande do Norte']
```

```python
dados_rn.head()
```

```python
# Os Motivos de indeferimento dos benefícios
dados_rn['Motivo Indeferimento'].value_counts().head(10).plot.barh()
```

```python
# Os tipos de Benefícios mais solicitados
dados_rn['Espécie'].value_counts().head(10).plot.barh()
```

```python
# Quem mais solicita o auxilio
dados_rn['Forma Filiação'].value_counts().head(10).plot.barh()
```

```python
# Categoria de quem pediu o auxilio
dados_rn['Clientela'].value_counts().head(10).plot.barh()
```
