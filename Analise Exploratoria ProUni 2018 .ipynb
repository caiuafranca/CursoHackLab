```python
#quantidade de cursos por UF
df['uf_busca'].value_counts(ascending=True).plot.barh()
```

```python
# Curso com a Maior Mensalidade
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# Curso com a Menos Mensalidade
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
#Cursos por UF
df_rn = df[df['uf_busca'] == 'RN']
```

```python
# Curso Prouni no Rn (Mais barato)
df_rn[df_rn['mensalidade'] == df_rn['mensalidade'].min()]
```

```python
#informações estatisticas do df
df_rn.describe()
```

```python
#Listagem por Grau do curso == tecnológico
df_rn[df_rn['grau'] == 'Tecnológico'][['nome','uf_busca','grau','mensalidade']]
```
